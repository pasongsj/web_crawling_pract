```python
from selenium import webdriver
from selenium.webdriver.common.by import By
NUM = 100
```

```python
def getStack(category):
    browser = webdriver.Chrome("/usr/local/bin/chromedriver")	#chromedriver 위치
    
    URL = 'https://www.wanted.co.kr/wdlist/518'+str(category)+'?country=kr&job_sort=company.response_rate_order&years=-1&locations=all'
    browser.get(URL);	# 페이지 열기
    browser.implicitly_wait(2)
    
    scroll = 1
    while(scroll):
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")
        infoList = browser.find_element(By.CSS_SELECTOR,'#__next > div.JobList_cn__t_THp > div.JobList_contentWrapper__QiRRW > div > ul')
        items = infoList.find_elements(By.TAG_NAME, 'li')
        if(len(items)>NUM):
            scroll = 0
            
    infoList = browser.find_element(By.CSS_SELECTOR,'#__next > div.JobList_cn__t_THp > div.JobList_contentWrapper__QiRRW > div > ul')
    items = infoList.find_elements(By.TAG_NAME, 'li') 
    
    
    

    pid = []
    num = 0
    for item in items[:NUM-4]:
        #섹션
        info = item.find_element(By.CSS_SELECTOR, 'div > a')
        
        #공고id
        idx = info.get_attribute('data-position-id')

        pid.append(idx)

    browser.quit()
    return pid
    
```

```python
getStack(872)
```
